```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df_train = pd.read_csv("Train.csv")
print(df_train.shape)
df_train.head()
```

```python
df_train.info()
```

```python
df_test = pd.read_csv("/content/Test.csv")
print(df_test.shape)
df_test.head()
```

```python
ss = pd.read_csv("/content/SampleSubmission.csv")
print(ss.shape)
ss.head()
```

```python
print(df_train.columns)
df_test.columns
```

```python
#To remove the commma from year
df_train['Year'] = df_train['Year'].str.replace(',','')
df_test["Year"] = df_test["Year"].str.replace(",", "")
df_test.head()
df_train.head()
```

```python
df_train['Year'] = pd.to_numeric(df_train['Year'], errors= "coerce", downcast= "integer")
df_test['Year'] = pd.to_numeric(df_test['Year'], errors= "coerce", downcast= "integer")

mean_value_year = df_train["Year"].mean()
df_train["Year"].fillna(mean_value_year, inplace=True)

test_mean_value_year = df_test["Year"].mean()
df_test["Year"].fillna(test_mean_value_year, inplace=True)
```

```python
df_train["Year"] = df_train["Year"].astype(int)
df_test["Year"] = df_test["Year"].astype(int)
```

```python
df_train.isna().sum()
df_test.isna().sum()
```

```python
#To fill missing values for string data tyoes
train_mode_value = df_train["Type"].mode()[0]
df_train["Type"].fillna(train_mode_value, inplace=True)

test_mode_value = df_test["Type"].mode()[0]
df_test["Type"].fillna(test_mode_value, inplace=True)
```

```python
mean_value_amount = df_train["Amount (Million Naira)"].mean()
df_train["Amount (Million Naira)"].fillna(mean_value_amount, inplace=True)

```

```python
df_train.isna().sum()
df_test.isna().sum()
```

```python
no_of_unique_dist = df_train["Distance"].unique()
len(no_of_unique_dist)
```

```python
df_train["Distance"] = pd.to_numeric(df_train["Distance"], errors= "coerce", downcast= "integer")

train_mean_value_distance = df_train["Distance"].mean()
df_train["Distance"].fillna(train_mean_value_distance, inplace=True)


test_mean_value_distance = df_test["Distance"].mean()
df_test["Distance"].fillna(test_mean_value_distance, inplace=True)

#to change distance data type to integer
df_train['Distance'] = df_train['Distance'].astype(int)
df_test['Distance'] = df_test['Distance'].astype(int)
```

```python
print(df_train.isna().sum())

df_test.isna().sum()
```

```python
df_train.drop(["VehicleID","Model", "Colour", "Distance"], axis=1, inplace=True)
df_test.drop(["VehicleID","Model", "Colour", "Distance"], axis=1, inplace=True)


```

```python
train_cols = ["Location", "Maker", "Year", "Type"]
for col in train_cols:
    print(col, "\n", df_train[col].unique())

test_cols = ["Location", "Maker", "Year", "Type"]
for col in test_cols:
    print(col, "\n", len(df_test[col].unique()))
```

```python
df_train = df_train.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
categorical_features = ["Location", "Maker", "Year", "Type"]
for col in categorical_features:
    labelencoder = LabelEncoder()
    df_train[col]= labelencoder.fit_transform(df_train[col])
    df_test[col]= labelencoder.fit_transform(df_test[col])

df_train.head()
```

```python
y = df_train["Amount (Million Naira)"]
X = df_train.drop("Amount (Million Naira)", axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr_pred = lr.predict(X_test)
np.sqrt(mean_squared_error(y_test, lr_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor() # Create the model
rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
np.sqrt(mean_squared_error(y_test, rf_pred))
```

```python
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(y_test, rf.predict(X_test)))
```
