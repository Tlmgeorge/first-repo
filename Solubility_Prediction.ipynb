# **ML Project to Predict solubility of Molecules**

# **Load Data**

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')

df
```

# **Data Preparation**

### **Data Seperation as X and y**

```python
y = df['logS']  #To view the logS data

y
```

```python
X = df.drop('logS', axis=1)  #To remove LogS column from the dataset

X
```

## **Data Splitting**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
X_train
```

```python
X_test
```

# **Model Building**

## **Linear Regression**

### **Training the Model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

### **Applying the Model to make a prediction**

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

### **Evaluating Model Performance**

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print("lr_train_mse: ", lr_train_mse)
print("lr_train_r2: ", lr_train_r2)
print("lr_test_mse: ", lr_test_mse)
print("lr_test_r2: ", lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results
```

```python
 lr_results.columns = ["Method", "Training Mse", "Training R2", "Test Mse", "Test R2"]
 lr_results
```

## **Random Forest**

### **Training the Model**

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

### **Applying the Model to Make a Prediction**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

### **Evaluating the Model**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results
```

```python
rf_results.columns = ["Method", "Training Mse", "Training R2", "Test Mse", "Test R2"]
rf_results
```

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```

## **Gradient Boost**

### **Training the Model**

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor(max_depth=2, random_state=100)
gb.fit(X_train, y_train)
```

### **Applying the Model to Make a Prediction**



```python
y_gb_train_pred = gb.predict(X_train)
y_gb_test_pred = gb.predict(X_test)
```

### **Evaluating the Model**

```python
from sklearn.metrics import mean_squared_error, r2_score

gb_train_mse = mean_squared_error(y_train, y_gb_train_pred)
gb_train_r2 = r2_score(y_train, y_gb_train_pred)

gb_test_mse = mean_squared_error(y_test, y_gb_test_pred)
gb_test_r2 = r2_score(y_test, y_gb_test_pred)
```

```python
gb_results = pd.DataFrame(['Gradient Boost', gb_train_mse, gb_train_r2, gb_test_mse, gb_test_r2]).transpose()
gb_results.columns = ["Method", "Training Mse", "Training R2", "Test Mse", "Test R2"]
gb_results
```

```python
df_models = pd.concat([lr_results, rf_results, gb_results], axis=0)
df_models
```
