# **Target Search**

```python
target = new_client.target
target_query = target.search("coronavirus")
targets_corona = pd.DataFrame.from_dict(target_query)
targets_corona
```

# **Selecting and Retrieving Data for SARS Coronavirus 3c-like proteinase**

```python
selected_corona_target = targets_corona.target_chembl_id[9]
selected_corona_target
```

```python
corona_activity = new_client.activity
corona_result = corona_activity.filter(target_chembl_id=selected_corona_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(corona_result)
```
