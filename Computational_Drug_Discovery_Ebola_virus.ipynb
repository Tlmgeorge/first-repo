# **Importing Libraries**

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

# **Target Search**

```python
target = new_client.target
target_query = target.search("ebolavirus")
targets_ebola = pd.DataFrame.from_dict(target_query)
targets_ebola
```

# **Selecting and Retrieving Data for Ebolavirus**

```python
selected_ebola_target = targets_ebola.target_chembl_id[0]
selected_ebola_target
```

```python
ebola_activity = new_client.activity
ebola_result = ebola_activity.filter(target_chembl_id=selected_ebola_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(ebola_result)
```
